&nbsp;K-means é um algoritmo de agrupamento popular usado em aprendizado de máquina não supervisionado. O algoritmo particiona um determinado conjunto de dados em k clusters, onde k é um número predeterminado.

&nbsp;O algoritmo começa selecionando aleatoriamente k pontos do conjunto de dados como centróides e, em seguida, atribui cada ponto de dados ao centróide mais próximo com base na métrica de distância usada (geralmente distância euclidiana). Após todos os pontos de dados terem sido atribuídos a um cluster, o algoritmo calcula a média de cada cluster para atualizar a posição do centróide. Este processo é repetido iterativamente até que os centróides não se movam mais significativamente, ou um número máximo de iterações tenha sido atingido.

&nbsp;O objetivo do algoritmo K-means é minimizar a soma das distâncias ao quadrado entre cada ponto de dados e seu centróide atribuído. Os clusters resultantes podem ser visualizados e interpretados para obter insights sobre a estrutura subjacente do conjunto de dados. K-means é amplamente utilizado em uma variedade de aplicações, incluindo segmentação de imagens, segmentação de mercado e detecção de anomalias.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

plt.style.use('default') 
```

```python
df = pd.read_csv("../Files/newdata.csv") 
df.info()
```

```python
scaler = StandardScaler()
data_1 = scaler.fit_transform(df.drop(["Model","Thrust"], axis=1))
data_2 = scaler.fit_transform(df.drop(["Model","Power"], axis=1))
```

### Avaliação da quantidade otimizada de grupos (k) para esse conjunto de dados   

```python
inertia = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++',
                    n_init='auto',random_state=42)
    kmeans.fit(data_2) 
    labels = kmeans.labels_
    inertia.append(kmeans.inertia_)
    score = silhouette_score(data_2, labels)
    print(f"k = {i} - score = {round(score,2)}")
    
plt.figure(figsize=(3,2))
plt.plot(range(2,11),inertia)
plt.show()
```

### Treinamento dos modelos 

```python
model_1 = KMeans(n_clusters=4, init ='k-means++', 
                  n_init='auto', random_state=42)
model_1.fit(data_1)
labels_1 = model_1.labels_

model_2 = KMeans(n_clusters=4, init ='k-means++', 
                  n_init='auto', random_state=42)
model_2.fit(data_2)
labels_2 = model_2.labels_
```

### Avaliação dos resultados

&nbsp;Para visualizar efetivamente o resultado de um modelo de agrupamento, é imperativo condensar todas as colunas da tabela em duas. Isso pode ser alcançado por meio de uma análise de componentes principais, uma técnica que permite a identificação de padrões subjacentes e correlações nos dados. Ao reduzir a dimensionalidade dos dados, o gráfico de dispersão resultante fornece uma representação simplificada dos pontos de dados e seus respectivos clusters, permitindo uma compreensão mais intuitiva do desempenho do modelo.

```python
#Transformação dos dados para duas dimensões
two_pca = PCA(n_components=2, random_state=42)
two_pca_vecs = two_pca.fit_transform(data_2)
x0 = two_pca_vecs[:, 0]
x1 = two_pca_vecs[:, 1]

#Criação de uma tabela contendo os resultados  
result_df = pd.DataFrame({"model1":labels_1,"model2":labels_2,"x0":x0,"x1":x1}, index=df.index)
result_df["predicted"] = "same"
result_df.loc[result_df.model1!=result_df.model2,"predicted"] = "diff"
result_df["Model"] = df.Model

#Visualização dos modelos
fig, ax = plt.subplots(1,3,figsize=(8,3), sharey=True)
sns.scatterplot(data=result_df, x="x0", y="x1", hue="model1", palette="viridis", ax=ax[0])
sns.scatterplot(data=result_df, x="x0", y="x1", hue="model2", palette="viridis", ax=ax[1])
sns.scatterplot(data=result_df, x="x0", y="x1", hue="predicted", style="predicted", 
                markers={"same":".","diff":"X"}, ax=ax[2])
ax[0].set_title("Com a variável antiga")
ax[1].set_title("Com a nova variável")
ax[2].set_title("Diferenças")
plt.suptitle("Comparação dos modelos de agrupamento")
plt.tight_layout()
plt.show()
```

```python
#Diferenças entre os dois modelos
result_df[result_df.predicted=="diff"]
```
