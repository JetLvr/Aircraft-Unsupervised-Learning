&nbsp;A tabela de origem apresenta várias questões que precisam ser abordadas para conduzir uma análise significativa. Para esse fim, é imperativo que preparemos os dados, identificando e corrigindo quaisquer inconsistências, imprecisões ou informações ausentes. Depois que tivermos Abordamos esses problemas, selecionaremos e transformaremos cuidadosamente os elementos relevantes nos formatos apropriados para análise, garantindo que nossas idéias sejam precisas e acionáveis .            

```python
import pandas as pd
import numpy as np
from string import ascii_letters

df = pd.read_json("../scrapy_project/output/data.jsonl", lines=True)
df.head(3)
```

```python
df[df.model=="McDonnell Douglas DC-8-72"]
```

```python
#Tratamento para os nomes das colunas
old_cols = ["".join(l for l in i if l in set(ascii_letters+" ")).split() 
            for i in df.columns] 
new_cols = ["".join(w.capitalize() for w in i) for i in old_cols]
str_df = df.rename(columns={k:v for k,v in zip(df.columns,new_cols)})

#Remoção do excesso de espaços e identificação dos valores nulos
str_df = str_df.astype(str).applymap(lambda x: " ".join(w for w in x.split())) 
str_df = str_df.applymap(lambda x: "None" if x == "" else x)
str_df = str_df.applymap(lambda x: np.nan if x == "None" else x)
```

```python
str_df.info()
```

```python
#Seleção das colunas avaliadas em relação ao número de valores nulos
str_df = str_df.iloc[:,[0,2,3,4,6,8,9,12,14,15,23,26]].dropna().reset_index(drop=True)
str_df.head(3)
```

&nbsp;As colunas "ExteriorLength" e "WingSpanRotorDiameter" contêm dois conjuntos de medições, um em metros e outro em pés. Devido a essa característica única, precisaremos tratá-los de maneira diferente das outras colunas da tabela. Especificamente, precisaremos separar os dois conjuntos de medições e convertê-los em uma unidade de medição padronizada para consistência e facilidade de análise.

```python
#Separação e tratamento da metade esquerda
le_df = str_df.iloc[:,2:-2]
le_df = le_df.applymap(lambda x: x.split()[0].replace(",","")).astype(float)

#Separação e tratamento da metade direita
ri_df = str_df.iloc[:,-2:]
ri_df = ri_df.applymap(lambda x: x.split()[-2]).astype(float)

#Junção das duas metades numa nova tabela
new_df = le_df.join(ri_df)

#Inserção dos modelos na nova tabela
new_df.insert(0,"Model",str_df.Model)

#Inserção do tipo de força do motor
power_unit = str_df.Power.map(lambda x: "lbf" if "pound-force" in x else "hp")
new_df.insert(2,"Unit",power_unit)
```

Algumas linhas da coluna "Engines" não contém a informação sobre quantos motores a aeronave possui.

```python
#Localização dos modelos onde o número de motores não é informado 
engines = str_df.Engine.map(lambda x: x.split()[0][0])
str_df.loc[engines[engines.str.isnumeric()==False].index,"Model"]
```

Essas quatro aeronaves possuem 2 motores cada.

```python
#Inserção da quantidade de motores na variável
engines[engines.str.isnumeric()==False] = "2"

#Inserção da variável com o número de motores na nova tabela 
new_df.insert(3,"EnginesQty",engines.astype(int))
```
